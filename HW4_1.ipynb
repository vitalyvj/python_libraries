### Задание 2

```python
t = np.linspace(0, 10, 51)
f = np.cos(t)

plt.plot(t, f, color='green')
plt.title('График f(t)', fontsize=15, fontweight='bold')
plt.xlabel('Значения t', fontsize=12)
plt.ylabel('Значения f', fontsize=12)
plt.xlim(0.5, 9.5)
plt.ylim(-2.5, 2.5)
plt.grid()
```

### Задание 3

```python
x = np.linspace(-3, 3, 51)
y1 = x ** 2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)

fig, ax = plt.subplots(nrows=2, ncols=2)
fig.set_size_inches(8, 6)
fig.subplots_adjust(wspace=0.3, hspace=0.3)

ax[0, 0].plot(x, y1)
ax[0, 1].plot(x, y2)
ax[1, 0].plot(x, y3)
ax[1, 1].plot(x, y4)

ax[0, 0].set_title('График y1')
ax[0, 1].set_title('График y2')
ax[1, 0].set_title('График y3')
ax[1, 1].set_title('График y4')

ax[0, 0].set_xlim(-5, 5)

plt.show()
```
