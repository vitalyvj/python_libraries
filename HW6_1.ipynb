```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import load_boston
```

```python
load_boston().keys()
```

```python
load_boston()['data'].shape
```

```python
X = pd.DataFrame(load_boston()['data'], columns=load_boston()['feature_names'])

X
```

```python
y = pd.DataFrame(load_boston()['target'], columns=['price'])

y
```

```python
X.info()
```

```python
y.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)

y_pred[:10]
```

```python
y_pred.shape
```

```python
check = pd.DataFrame({'y_test': y_test['price'], 'y_pred': y_pred.flatten()})

check
```

```python
check['diff'] = check['y_test'] - check['y_pred']
check['diff_%'] = check['diff'] / check['y_test'] * 100

check.sort_values('diff_%')
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(check['y_test'], check['y_pred'])
```

### Задание 4

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV, train_test_split
```

```python
df = pd.read_csv('datasets/creditcard.csv')

df
```

```python
df['Class'].value_counts(normalize=True)
```

```python
df.info()
```

```python
pd.options.display.max_columns = 100
```

```python
df
```

```python
X = df.drop('Class', axis=1)

X
```

```python
y = df['Class']

y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100, stratify=y)
```

```python
print(*(i.shape for i in (X_train, y_train, X_test, y_test)), sep='\n')
```

```python
parameters = [{
    'n_estimators': [10, 15],
    'max_features': np.arange(3, 5),
    'max_depth': np.arange(4, 7)
}]
```

```python
clf = GridSearchCV(
    estimator=RandomForestClassifier(random_state=100),
    param_grid=parameters,
    scoring='roc_auc',
    cv=3
)
```

```python
clf.fit(X_train, y_train)
```

```python
clf.best_params_
```

```python
y_pred_proba = clf.predict_proba(X_test)[:, 1]

y_pred_proba
```

```python
from sklearn.metrics import roc_auc_score
```

```python
roc_auc_score(y_test, y_pred_proba)
```
